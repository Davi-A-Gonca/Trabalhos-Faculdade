```python
import random

def funcao_Pedra_Papel_Tesoura(opcao_jogador):
  escolhas_possiveis = ["Pedra", "Papel", "Tesoura", "Lagarto", "Spock"]
  opcao_computador = random.randint(1,5)

  print(f"\nVocê escolheu: {escolhas_possiveis[opcao_jogador - 1]}")
  print(f"O computador escolheu: {escolhas_possiveis[opcao_computador - 1]}\n")

  if opcao_jogador == opcao_computador:
    return "Empate!"
  elif (opcao_jogador in (3, 4) and opcao_computador == 1) or \
       (opcao_jogador in (1, 5) and opcao_computador == 2) or \
       (opcao_jogador in (2, 4) and opcao_computador == 3) or \
       (opcao_jogador in (2, 5) and opcao_computador == 4) or \
       (opcao_jogador in (1, 3) and opcao_computador == 5):
    return "Você perdeu ahahaha otário! Quero dizer, que pena...."
  elif (opcao_computador in (3, 4) and opcao_jogador == 1) or \
       (opcao_computador in (1, 5) and opcao_jogador == 2) or \
       (opcao_computador in (2, 4) and opcao_jogador == 3) or \
       (opcao_computador in (2, 5) and opcao_jogador == 4) or \
       (opcao_computador in (1, 3) and opcao_jogador == 5):
    return "Você Ganhou... Quero dizer, você ganhou!!"


print("Vamos jogar Pedra-Papel-Tesoura");

while True:
  opcao = int(input("Qual sua escolha?\n[1] Pedra\n[2] Papel\n[3] Tesoura\n[4] Lagarto\n[5] Spock\n[6] Sair\n"))

  if opcao == 6:
    print("Adeus, obrigado por jogar!")
    break
  elif opcao in (1, 2, 3, 4, 5):
    print(funcao_Pedra_Papel_Tesoura(opcao));
  else:
    print("Opção inválida, seu feio. Tente de novo\n")


```
